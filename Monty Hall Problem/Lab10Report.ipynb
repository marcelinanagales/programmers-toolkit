## Exercise 3

```python
data = np.loadtxt("afile.txt")
```

```python
m_mesh = np.linspace(3, 100, 100);
n_mesh = np.linspace(1, 98, 100)

m = data[:, 0]
n = data[:, 1]
```

```python
z = data[:, 2]
M, N = np.meshgrid(m_mesh, n_mesh)
rbf = scipy.interpolate.griddata((m, n), z,(M, N), method='linear')
```

```python
fig=plt.contour(m_mesh, n_mesh, rbf)
plt.clabel(fig, inline=1, fontsize=10)
CB = plt.colorbar(fig, shrink=0.8, extend='both')
plt.xlabel("Number of Doors")
plt.ylabel("Number of Revealed Empty Doors")
plt.title("Contour plot of probability of winning")
plt.show()
```

```python
fig2=plt.contour(m_mesh, n_mesh, rbf)
plt.scatter(m, n, c=z)
CB = plt.colorbar(fig2, shrink=0.8, extend='both')
plt.xlabel("Number of Doors")
plt.ylabel("Number of Revealed Empty Doors")
plt.title("Contour plot of probability of winning")
plt.show()
```

Above are the two contour plots of the probability of winning, given the number of doors (m) and the number of empty doors revealed before the player switches their choice. Both graphs show that the closer the number of doors revealed to the amount of doors there are, there is a greater probability of winning, and even then, the chances are slim.
<br><br>
The rest of the graph, besides those strong lines near slope =1, has a very low chance of winning, even if the player switches doors, which is logical.
